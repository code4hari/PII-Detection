```python
import pandas as pd

# Load the dataset
df = pd.read_csv('sample_data.csv')

# Display the first few rows of the dataset
print(df.head())

# Summary statistics
print(df.describe())

# Check for missing values
print(df.isnull().sum())

# Check data types
print(df.dtypes)

# Data preprocessing
# Remove any leading or trailing whitespaces in columns
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

# Convert phone numbers to a consistent format (e.g., remove special characters)
df['phone_number'] = df['phone_number'].str.replace(r'\D', '')

# Convert email addresses to lowercase
df['email'] = df['email'].str.lower()

# Remove any leading or trailing whitespaces in the address column
df['address'] = df['address'].str.strip()

# Convert SSN to a consistent format (e.g., remove special characters)
df['ssn'] = df['ssn'].str.replace(r'\D', '')

# Display the updated DataFrame
print(df.head())
```
